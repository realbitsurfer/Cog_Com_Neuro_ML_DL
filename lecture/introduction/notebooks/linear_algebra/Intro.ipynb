## Linear algebra - what was that again?

Linear algebra is the math of arrays (of numbers) and not too strongly related to actual algebra, which deals with equations and unknown variables. In linear algebra we deal with objects which are contained in the linear space, such as vectors and matrices, and perform mathematical operations on them. You could also call linear algebra **the math of data**, since its objects resemble the containers in which data is stored in.
